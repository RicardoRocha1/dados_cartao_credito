```python
import pandas as pd

excel_file_path = 'excels/Demanda-das-Empresas-por-Credito-Maio2024.xlsx'
credito_df = pd.read_excel(excel_file_path, sheet_name='Var. Mensal', skiprows=5, usecols='AC:AJ', header=None, index_col= None, names=['Industria', 'Comercio', 'Servicos', 'Demais', 'MPE', 'Medias', 'Grandes',	'Total'])

index_df = pd.read_excel(excel_file_path, sheet_name='Var. Mensal', skiprows=5, usecols='A', header=None)
index_series = index_df.iloc[:, 0]
credito_df.set_index(index_series, inplace=True)
credito_df.index = credito_df.index.astype(str)
credito_df.index.name = 'Data'

parquet_file_path = 'parquet/demanda_credito_cnpj.parquet'
credito_df.to_parquet(parquet_file_path, engine='pyarrow')
credito_df.head()

# Os dados aqui contidos possuem indicadores tanto por setor (Indústria,...,Demais) e por tamanho de empresa (MPE, Médias, Grandes). Isto será tratado posteriormente para manter uma entrada com os dados brutos.
```
