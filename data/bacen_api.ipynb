```python
import requests
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.types import StructType, StructField, StringType, FloatType, ArrayType
from pyspark.sql.functions import col
import pandas as pd

# A URL abaixo já está configurada para os dados a serem extraidos, é possivel modificar os parametros caso necessário.
base_url = 'https://olinda.bcb.gov.br/olinda/servico/taxaJuros/versao/v2/odata/TaxasJurosDiariaPorInicioPeriodo?$top=10000&$format=json&$select=InicioPeriodo,FimPeriodo,Segmento,Modalidade,InstituicaoFinanceira,TaxaJurosAoMes'
```

```python
response = requests.get(url=base_url)
response = response.json()
```

```python
response
```

```python
spark = SparkSession.builder.appName('bacen_api').getOrCreate()

bacen_cartao_credito = spark.createDataFrame([response])
bacen_cartao_credito.show(truncate=False)
```
